```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')
df
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax = sns.scatterplot(data=df, x='Year', y='CSIRO Adjusted Sea Level')

x, y = df['Year'], df['CSIRO Adjusted Sea Level']
lin_reg = linregress(x, y)
# print(lin_reg.slope, lin_reg.intercept)
x_pred = list(range(1880, 2050))
y_pred = [lin_reg.slope * year + lin_reg.intercept for year in x_pred]

ax.plot(x_pred, y_pred, color='red')

df_new = df[df['Year'] >= 2000]
x_new, y_new = df_new['Year'], df_new['CSIRO Adjusted Sea Level']
lin_reg_new = linregress(x_new, y_new)
x_pred_new = list(range(2000, 2051))  
y_pred_new = [lin_reg_new.slope * year + lin_reg_new.intercept for year in x_pred_new]

ax.plot(x_pred_new, y_pred_new, color='green')
ax.set_xlabel('Year')
ax.set_ylabel('Rise in Sea Level')
```
